```python
import pandas as pd

#Reads the data
rawdata = pd.read_csv('./Demanda_2015.txt', sep=';', encoding='utf-8')
```

```python
rawdata.head()
```

```python
import matplotlib.pyplot as plt

#groups data per day (header is date;time;consumption where date is the dd/mm/yy and time is the hour)
unique_times = rawdata['time'].unique().tolist()

data_per_day = rawdata.groupby('date')

#print(data.head())
plt.figure(figsize=(14, 8))
for day, group in data_per_day:
    plt.plot(group['time'], group['consumption'], label=day)

plt.xticks(range(0, len(unique_times), 6), unique_times[::6], rotation=45)
#x label
plt.xlabel('Time')
#y label
plt.ylabel('Consumption')
```
