```python
from RTA import RTA
model = RTA("map.txt")
```

```python
model.hard_reset()
print(model)
prev_pos_str = model.map.position()
prev_x, prev_y = model.map.pos
prev_h = model.map.h[prev_x][prev_y]

while model.forward():
    current_h = model.map.h[prev_x][prev_y]
    current_pos_str = model.map.position()

    print(f"Paso {model.steps}: h({prev_pos_str}) = {prev_h} -> {current_h}. Se mueve a la casilla {current_pos_str}.")
    print(model)

    prev_x, prev_y = model.map.pos
    prev_h = model.map.h[prev_x][prev_y]
    prev_pos_str = current_pos_str

current_h = model.map.h[prev_x][prev_y]
current_pos_str = model.map.position()

print(f"Paso {model.steps}: h({prev_pos_str}) = {prev_h} -> {current_h}. Se mueve a la casilla {current_pos_str}.")
#print(model)

prev_pos_str = model.map.position()
prev_h = current_h

print(f"Se ha llegado al objetivo. Los pasos realizados son: {model.actions}.")
print(f"La ruta es: {model.path}")
```
